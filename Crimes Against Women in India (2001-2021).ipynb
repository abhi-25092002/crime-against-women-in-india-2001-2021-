# VISULIZATION:-

# 1. WHICH AREA HIGHEST CRIMES AGAINST WOMEN IN INDIA

```python
dt['State'].value_counts()[:].plot(kind='bar',color='purple',figsize=(25,10),title=('Which Area High Crime In India'))
```

# 2.WHICH YEAR HIGHEST CRIMES AGAINST WOMEN IN INDIA

```python
dt['Year'].plot(kind='line',color='r',figsize=(25,10),title=('Which Year High Crime In India')).grid(color='black',linestyle='dotted',linewidth=0.5)
```

# 3.CHECKING CRIME YEAR REGARDING AREA

```python
plt.subplots(figsize=(15,25))
sns.boxplot(y='State',x='Year',data=dt,palette='rainbow')
plt.show
```

```python
dt.head(2)
```

# 4.CHECKING STATE ACCORDING RAPE CASES

```python
plt.subplots(figsize=(15,15))
sns.barplot(y='State',x='Rape',data=dt)
plt.show
```

```python
dt.head()
```

# 5.YEAR REGARDING KIDNAP AND ASSULT CASES

```python
plt.subplots(figsize=(15,5))
sns.boxplot(y='K&A',x='Year',data=dt)
plt.title('YEAR AND K&A CASES',fontsize=15,color='olive')
plt.show
```

# 6.CHECKING STATE WISE AoW CRIMES

```python
plt.figure(figsize=(15,15))
sns.scatterplot(x='AoW',y='State',data=dt)
plt.xlabel('Aow',size=20)
plt.ylabel('State',size=20)
```

# 7.YEAR ACCORDING WT AND DV CRIMES AGAINST WOMEN IN INDIA

```python
sns.lmplot(x='WT',y='Year',data=dt)
plt.xlabel('WT',size=15)
plt.ylabel('Year',size=15)
```

```python
dt.head()
```

```python
plt.subplots(figsize=(15,8))
sns.barplot(x='Year',y='DV',data=dt)
plt.title('WHICH YEAR HIGHEST DV CRIMES',color='g',fontsize=15)
plt.show
```
